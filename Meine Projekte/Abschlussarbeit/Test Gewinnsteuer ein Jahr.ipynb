#### Ich putze den Datensatz (diese Strichli sind mir ein Dorn im Auge)

```python
def geputzt(zeile):
    zeile = zeile.replace('(', '')
    zeile = zeile.replace(')', '')
    zeile = zeile.replace("'", "").strip()
    return zeile
```
