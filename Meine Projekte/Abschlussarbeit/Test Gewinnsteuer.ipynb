#### Ich putze den Datensatz (diese Strichli sind mir ein Dorn im Auge)

```python
def geputzt(zeile):
    zeile = zeile.replace('(', '')
    zeile = zeile.replace(')', '')
    zeile = zeile.replace("'", "").strip()
    return zeile
```

```python
firmen_2007['Steuerbarer Gewinn ab CHF'] = firmen_2007['Steuerbarer Gewinn ab CHF'].apply(geputzt)
```

```python
firmen_2007['Anzahl Steuerpflichtige'] = firmen_2007['Anzahl Steuerpflichtige'].apply(geputzt)
firmen_2007['Betrag Kantonssteuer CHF'] = firmen_2007['Betrag Kantonssteuer CHF'].apply(geputzt)
firmen_2007
```
