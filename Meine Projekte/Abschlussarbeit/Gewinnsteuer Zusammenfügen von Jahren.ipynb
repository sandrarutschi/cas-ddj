Mein Code, verkürzt, ausführlich in Test Gewinnsteuer ein Jahr

```python
#zurechtschneiden
test_df = Firmen_2007[0].df
Firmen_07 = test_df[12:]
firmen_2007 = Firmen_07.head(5)

#headers setzen
headers = ['Weglassen', 'Steuerbarer Gewinn ab CHF', 'weglassen', 'Anzahl Steuerpflichtige 2007', 'Betrag Kantonssteuer 2007 in CHF', 
           'lassweg']
firmen_2007.columns = headers

#überflüssige kolonnen weg
del firmen_2007['Weglassen']
del firmen_2007['weglassen']
del firmen_2007['lassweg']

#putzen
def geputzt(zeile):
    zeile = zeile.replace('(', '')
    zeile = zeile.replace(')', '')
    zeile = zeile.replace("'", "").strip()
    return zeile
firmen_2007['Steuerbarer Gewinn ab CHF'] = firmen_2007['Steuerbarer Gewinn ab CHF'].apply(geputzt)
firmen_2007['Anzahl Steuerpflichtige 2007'] = firmen_2007['Anzahl Steuerpflichtige 2007'].apply(geputzt)
firmen_2007['Betrag Kantonssteuer 2007 in CHF'] = firmen_2007['Betrag Kantonssteuer 2007 in CHF'].apply(geputzt)

#um 90 Grad drehen
firmen_07 = firmen_2007.T

#headers weg
Firmen07 = firmen_07[1:]

#headers neu setzen
headers = ['Steuerbarer Gewinn 0-10000 CHF', 'Steuerbarer Gewinn 10100-50000 CHF', 'Steuerbarer Gewinn 50100-250000 CHF', 
           'Steuerbarer Gewinn ab 250100 CHF', 'Total Steuerbarer Gewinn']
Firmen07.columns = headers

#nur den Betrag Kantonssteuer rausziehen
Betrag_Kantonssteuer_2007 = Firmen07.drop(['Anzahl Steuerpflichtige 2007'])
Betrag_Kantonssteuer_2007
```

```python
# Anzahl Steuerpflichtige rausziehen
Anzahl_Steuerpflichtige_2007 = Firmen07.drop(['Betrag Kantonssteuer 2007 in CHF'])
Anzahl_Steuerpflichtige_2007
```
