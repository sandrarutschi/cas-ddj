# Einkommen Zusammenfügen von Jahren

```python
import camelot
import tabula
import pandas as pd
```

## 2007

```python
Einkommenssteuer_2007 = camelot.read_pdf("Einkommen_2007.pdf", flavor='stream', pages='all', table_areas=['0,842,595,0'])
```

```python
#zurechtschneiden
test_df = Einkommenssteuer_2007[0].df
Einkommen_07 = test_df[9:]
einkommen_2007 = Einkommen_07.head(18)

# Headers setzen
headers = ['Einkommen von bis', 'Anzahl Alleinstehende 2007', 'EK Alleinstehende', 
           'Betrag Kantonssteuer Alleinstehende 2007', 'Anzahl Ehepaare 2007', 'EK Ehepaare', 
           'Betrag Kantonssteuer Ehepaare 2007', 'Total Anzahl Steuerzahler', 'Total EK', 
           'Total Betrag Kantonssteuer']
einkommen_2007.columns = headers

# überflüssige Kolonnen löschen
del einkommen_2007['EK Alleinstehende']
del einkommen_2007['EK Ehepaare']
del einkommen_2007['Total EK']
del einkommen_2007['Total Betrag Kantonssteuer']
del einkommen_2007['Total Anzahl Steuerzahler']

# aus einer Kolonne zwei Kolonnen machen, Headers setzen und die überflüssige Kolonne löschen
getrennte_kolonnen07 = einkommen_2007['Einkommen von bis'].str.split('\n', 1, expand=True) 
headers = ['Einkommen ab CHF', 'streichen']
getrennte_kolonnen07.columns = headers
del getrennte_kolonnen07['streichen']

#die beiden df einkommen_2007 und getrennte_kolonnen07 zusammenfügen
Einkommen_2007_neu = pd.concat([getrennte_kolonnen07, einkommen_2007], axis=1)

# überflüssige Kolonne weg
del Einkommen_2007_neu['Einkommen von bis']

#putzen
def geputzt(zeile):
    zeile = zeile.replace('(', '')
    zeile = zeile.replace(')', '')
    zeile = zeile.replace('\n', '')
    zeile = zeile.replace("'", "").strip()
    return zeile
Einkommen_2007_neu['Einkommen ab CHF'] = Einkommen_2007_neu['Einkommen ab CHF'].apply(geputzt)
Einkommen_2007_neu['Anzahl Alleinstehende 2007'] = Einkommen_2007_neu['Anzahl Alleinstehende 2007'].apply(geputzt)
Einkommen_2007_neu['Betrag Kantonssteuer Alleinstehende 2007'] = Einkommen_2007_neu['Betrag Kantonssteuer Alleinstehende 2007'].apply(geputzt)
Einkommen_2007_neu['Anzahl Ehepaare 2007'] = Einkommen_2007_neu['Anzahl Ehepaare 2007'].apply(geputzt)
Einkommen_2007_neu['Betrag Kantonssteuer Ehepaare 2007'] = Einkommen_2007_neu['Betrag Kantonssteuer Ehepaare 2007'].apply(geputzt)

# Tabelle um 90 Grad drehen
Einkommen_2007_neu = Einkommen_2007_neu.T

# headers neu
headers = ['Einkommen 0 CHF', 'Einkommen 100-20000 CHF', 'Einkommen 20100-40000 CHF', 'Einkommen 40100-60000 CHF',
           'Einkommen 60100-80000 CHF', 'Einkommen 80100-100000 CHF', 'Einkommen 100100-125000 CHF', 
           'Einkommen 125100-150000 CHF', 'Einkommen 150100-175000 CHF', 'Einkommen 175100-200000 CHF',
           'Einkommen 200100-250000 CHF', 'Einkommen 250100-300000 CHF', 'Einkommen ab 300100-400000 CHF',
           'Einkommen 400100-500000 CHF', 'Einkommen 500100-750000 CHF', 'Einkommen 750100-1000000 CHF',
           'Einkommen ab 1000100 CHF', 'Total']
Einkommen_2007_neu.columns = headers

# erste Zeile weg
Einkommen07 = Einkommen_2007_neu[1:]

# Objekte in Integers verwandeln, damit ich die Zahlen addieren kann
Einkommen07 = Einkommen07.astype(int)

# Die einzelnen Zeilen rausziehen
Betrag_Alleinstehende07 = Einkommen07.drop(['Anzahl Ehepaare 2007', 'Anzahl Alleinstehende 2007',
                                            'Betrag Kantonssteuer Ehepaare 2007'])

Anzahl_Alleinstehende07 = Einkommen07.drop(['Anzahl Ehepaare 2007', 'Betrag Kantonssteuer Alleinstehende 2007',
                                            'Betrag Kantonssteuer Ehepaare 2007'])

Anzahl_Ehepaare07 = Einkommen07.drop(['Anzahl Alleinstehende 2007', 'Betrag Kantonssteuer Alleinstehende 2007',
                                      'Betrag Kantonssteuer Ehepaare 2007'])

Betrag_Ehepaare07 = Einkommen07.drop(['Anzahl Ehepaare 2007', 'Anzahl Alleinstehende 2007',
                                      'Betrag Kantonssteuer Alleinstehende 2007'])

Einkommen07
```
