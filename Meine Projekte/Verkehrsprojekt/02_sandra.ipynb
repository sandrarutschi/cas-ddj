### Und nun hängen wir alls zusammen

```python
#um das nicht für jede einzeln zu machen, bauen wirs in einen for-loop ein
#eine leere df bauen
ff = pd.DataFrame()
headers = ['Gemeinde', 'Sachschaden', 'Personenschaden', 'Getötete', 
           'Schwerverletzte', 'Lebensbedrohlich Verletzte', 
           'Erheblich Verletzte', 'Leichtverletzte']

#Und nun durch alles iterieren
#https://pandas.pydata.org/pandas-docs/stable/user_guide/merging.html
for number in range(0,8):
    temp = tables_stream[number].df #hier sagen wir, immer vorwärts gehen zur nächsten seite
    temp = temp[9:] #hier sagen wir: aber im fall erst aber der 9. reihe
    temp.columns = headers #und dann setz uns bitte noch diese headers, die wir da definitert haben
    frames = [ff, temp]  #so kleben wir eine tabelle an die andere an. ff ist ein neues frame, temp ist unsere einzelne tabelle oben
    ff = pd.concat(frames)   #mit concat kommts dann in die richtige form
```

```python
ff.info()
```

```python
## Mit leeren Strings umgehen, das dataframe putzen 
def clean(elem):
    if elem == "": #wenn das element ein leerer string ist, gibt mir 0 zurück. sonst gibt mir das element zurück
        return 0
    else:
        return elem
ff = ff.applymap(clean) #apply kann man für einzelne kolonnen machebn, applymap ist für das gesagte frame. das was in klammern kommt, machen wir dann
```

```python
## Die Gemeinden zwischenzeitlich löschen und dann wieder reinnehmen bei der Umwandlung in Zahlen. ff ist das gesamte dataframe
gem = ff['Gemeinde'] 
del ff['Gemeinde'] #damit schneide ich die erste kolonne ab
ff = ff.astype(int) #wir wandeln alle strings um und definieren sie als integer
ff['Gemeinde'] = gem #danach wende ich das wieder an, das ich gelöscht habe
ff = ff[['Gemeinde', 'Sachschaden', 'Personenschaden', 'Getötete', 
           'Schwerverletzte', 'Lebensbedrohlich Verletzte', 
           'Erheblich Verletzte', 'Leichtverletzte']].copy() ##Reihenfolge der Headers kann ich so definieren, ist eigentlich nur kosmetik
```

```python
#index anpassen
ff = ff.reset_index() #so läuft das von 1 an durch dann, icht erst ab 9
del ff['index'] # und so lösche ich dann die index-kolonne, die index heisst und nun nicht mehr nötig ist
```
