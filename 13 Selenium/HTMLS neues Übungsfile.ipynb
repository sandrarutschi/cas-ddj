```python
def extractbakeries(elem):
    zeilen= h.find_all('tr', {'class':'ng-scope'})
    blist = []
    
    for zeile in zeilen:
        name = zeile.find("div").text
        u_id = zeile.find_all('div')[2].text
        sitz = zeile.find_all('div')[4].text
        kanton = zeile.find_all('div')[5].text
        url = zeile.find('div').find_next('div').find('a')['href']
        
        minidict = {'Name': name,
           'u_id' : u_id,
           'Sitz' : sitz,
           'Kanton' : kanton,
           'url': url}
    
        blist.append(minidict)
    return blist

#so definiere ich eine neue funktion, die ich immer wieder will. den extrakt dieser bäckereien auf der ersten seite. 
#ich definiere die funktion, weil ich da ja dann wieder durchloopen muss und es sonst sehr unübersichtlich wäre. 
#so muss ich nicht nochmals den ganzen code reinschreiben im neuen loop, sondern nur die funktion
```

```python
finallist = []

for filename in lst:
    
    file = open('pages/' + filename, 'r')
    text = file.read()
    h = BeautifulSoup(text, 'html.parser')
    
    finallist = finallist + extractbakeries(h) #ich appende das nicht mit .append, sondern setze es einfach an.
    #weil sonst wäre es eine liste in der liste, nicht eine ganze neue liste
```
