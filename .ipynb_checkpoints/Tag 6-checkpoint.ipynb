```python
# hier nun im for-loop:
#ich definieren diese varialben:
#das hier ist das gerüst, das wir für grosse abfragen brauchen. 

urlanfang = ('https://earthquake.usgs.gov/fdsnws/event/1/count?starttime=2014-09-0') #den Anfang ohne Enddatum nehmen
string1 = 1
urlmitte = ("&endtime=2014-09-0") #den zweiten teil ohne enddatm
string2 = 2
newlst=[]

for strings in range(0,3):  #der range sagt, ich gehe dreimal durch
    print(urlanfang + str(string1) + urlmitte + str(string2))
    string1 = string1 + 1  #damit sorge ich wieder dafür, dass es vorwärts zählt)
    string2 = string2 + 1  #ich muss das für beide variablen machen
    r = requests.get(urlanfang + str(string1) + urlmitte + str(string2))
    print(r.text)
    newlst.append(r.text)
```
