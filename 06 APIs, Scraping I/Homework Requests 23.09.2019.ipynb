4. Zeige eine Liste der 10 st채rksten Erdbeben des laufenden Jahres an. (Nicht vergessen, nun musst Du die Methode wechseln. Nicht mehr "count", sondern "query").

```python
r = requests.get('https://earthquake.usgs.gov/fdsnws/event/1/query?format=geojson&starttime=2019-01-01&endtime=2019-09-23&minmagnitude=5.5')
die_liste = r.json()
```

```python
features_liste = die_liste["features"]
mag_liste = []

for key in features_liste:
    mag = key["properties"]["mag"]
    mag_liste.append(mag)
    
print(mag_liste)
```

5. Wo haben sich diese Erdbeben ereignet?

```python
features_liste = die_liste["features"]
ort_mag_liste = []

for key in features_liste:
    if key["properties"]["mag"]>=6.9:
        mag = key["properties"]["mag"]
        ort = key["properties"]["place"]
        mini_dict = {"magnitude":mag,
                    "place":ort}
        ort_mag_liste.append(mini_dict)
    
print(ort_mag_liste)
```

6. Ziehe auch die Zeit mit heraus?

```python
features_liste = die_liste["features"]
zeit_ort_mag_liste = []

for key in features_liste:
    if key["properties"]["mag"]>=6.9:
        mag = key["properties"]["mag"]
        ort = key["properties"]["place"]
        zeit = key["properties"]["time"]
        mini_dict = {"magnitude":mag,
                    "place":ort,
                    "zeit":zeit}
        zeit_ort_mag_liste.append(mini_dict)
    
print(zeit_ort_mag_liste)
```

```python
pd.DataFrame(zeit_ort_mag_liste)
```
