```python
df[['CS GROUP','UBS GROUP']].plot()
```

```python
df[['CS GROUP','UBS GROUP']].head(1)
```

```python
df[['CS GROUP','UBS GROUP']].loc['2014-10-13']
```

```python
erstezeile = df[['CS GROUP','UBS GROUP']].loc['2014-10-13']
```

```python
df[['CS GROUP','UBS GROUP']] / erstezeile * 100
```

```python
dfangeglichen = df[['CS GROUP','UBS GROUP']] / erstezeile * 100
```

```python
dfangeglichen.plot()
```

```python
#Die UBS legte im ersten Halbjahr 2015 stark an Wert zu. Den Vorsprung gegenüber CS konnte sie bis Ende 2017 weiterhin ausbauen. Seither sinkt der Wert jedoch, während CS im Vergleich konstanter bleibt. 
```

```python
erstezeilezwei = df[['ROCHE','NOVARTIS']].loc['2014-10-13']
```

```python
dfangeglichenzwei = df[['ROCHE','NOVARTIS']] / erstezeilezwei * 100
```

```python
dfangeglichenzwei.plot()
```
